```python
def get_response(url,key):
    print('正在抓取{}页面广告'.format(key))
    n_key=quote(key)
    n_url=url.format(n_key)
    r=requests.get(n_url,headers=headers)
    if r.status_code==200:
        return r.text
    else:
        print('正在重试')
        time.sleep(60)
        return get_response(url,key)
    
    
    
```

```python
def parse_response(key,res):
    print('正在解析{}页面'.format(key))
    c=0
    result=pd.DataFrame()
    all_key=[]
    all_host=[]
    all_time=[]
    all_title=[]
    all_red_title=[]
    all_content=[]
    all_red_content=[]
    all_rank=[]
    
    doc=pq(res)
    items=doc('div#content_left').children().items()
    for item in items:
        ##\33 001 > div.c-abstract.cuEeGN.WaYGLv > div > div.c-span18.c-span-last > div:nth-child(1) > font:nth-child(1)
        c+=1
       
        a=item.find('a>span').text()
        process_a=a.split(' ')
        #print(a)
        try:
            host=process_a[0]
        except Exception:
            host=''
        host=host.encode('gbk','ignore').decode('gbk','ignore')
        if '广告' in process_a:
            
        
           
            all_host.append(host)
            time = datetime.datetime.now().strftime('%Y/%m/%d %H:%M')
            all_time.append(time)
            title=item.find('h3 > a').text()
            title=title.encode('gbk','ignore').decode('gbk','ignore')
            all_title.append(title)
            red_title=item.find('h3>a>em').text()
            red_title=red_title.encode('gbk','ignore').decode('gbk','ignore')
            all_red_title.append(red_title)
            content=item.find('div.c-abstract').text()
            content=content.encode('gbk','ignore').decode('gbk','ignore')
            all_content.append(content)
            red_content=item.find('div.c-abstract>div>div>div>font').text()
            red_content=red_content.encode('gbk','ignore').decode('gbk','ignore')
            all_red_content.append(red_content)
            all_key.append(key)
            
        else:
            
            
            all_host.append(host)
            time = datetime.datetime.now().strftime('%Y/%m/%d %H:%M')
            all_time.append(time)
            title=item.find('h3 > a').text()
            title=title.encode('gbk','ignore').decode('gbk','ignore')
            all_title.append(title)
            red_title=item.find('h3>a>em').text()
            red_title=red_title.encode('gbk','ignore').decode('gbk','ignore')
            all_red_title.append(red_title)
            content=item.find('div.c-abstract').text()
            content=content.encode('gbk','ignore').decode('gbk','ignore')
            all_content.append(content)
            red_content=item.find('div.c-abstract>em').text()
            red_content=red_content.encode('gbk','ignore').decode('gbk','ignore')
            all_red_content.append(red_content)
            all_key.append(key)
        all_rank.append(c)
        
    result['抓取时间']=all_time
    result['搜索词']=all_key        
    result['创意标题']=all_title
    result['标题飘红']=all_red_title
    result['创意描述']=all_content
    result['描述飘红']=all_red_content
    result['广告主']=all_host
    result['排名']=all_rank
    print(result)
    return result
```

```python
res=get_response(base_url,'礼品鲜花')
parse_response('礼品鲜花',res)
```

```python
def load_data(file):
    data=pd.read_csv(file,encoding='gbk')
    return data
def process(data):
    keywords=data['关键词']
    nk=keywords.map(lambda x: '江苏'+x)
    return nk
```

```python
def save(outputfile,result_list):
    result=pd.concat(result_list,ignore_index=True)
    result.to_csv(outputfile,index=False,encoding='gbk')
```

```python
file='鲜花礼品.csv'
outputfile='result.csv'
data=load_data(file)
nk=process(data)
length=len(nk)
l=[]
for i in range(length):
    key=nk[i]
    res=get_response(base_url,key)
    r=parse_response(key,res)
    l.append(r)
save(outputfile,l)

```
