```python
import requests
from bs4 import BeautifulSoup
```

```python
#获取url下的页面内容，返回soup对象
def get_page(url):
    responce=requests.get(url)
    soup=BeautifulSoup(responce.text,'lxml')
    return soup
#获取列表页下面的所有租房页面下面的链接，返回一个链接列表
def get_links(url):
    soup=get_page(url)
    links_div=soup.find_all("div",class_="content__list--item--main")
    links=[("https://bj.lianjia.com"+ div.a.get('href')) for div in links_div]
    return links
```

```python
def get_information(url):
    soup=get_page(url)
    price=soup.find('p',class_='content__aside--title').text.strip()
    a=soup.find_all('p',class_='content__article__table')
    house_info=a[0].text.strip().split('\n')
    house=house_info[0]
    typ=house_info[0]
    area=house_info[2]
    orient=house_info[3]
    return price,house,typ,area,orient
```

```python
price_sum=[]
typ_sum=[]
area_sum=[]
orient_sum=[]
for i in range(30):
    house_url=b[i]
    soup=get_page(house_url)
    price=soup.find('p',class_='content__aside--title').text.strip()
    a=soup.find_all('p',class_='content__article__table')
    house_info=a[0].text.strip().split('\n')
    house=house_info[0]
    typ=house_info[0]
    area=house_info[2]
    orient=house_info[3]
    price_sum.append(price)
    typ_sum.append(typ)
    area_sum.append(area)
    orient_sum.append(orient)
    print(price_sum,typ_sum,area_sum)
    
```

```python
result={}
result['price']=price_sum
result['area']=area_sum
result['orient']=orient_sum
```
